```python
import schemdraw
from schemdraw import pictorial
import schemdraw.elements as elm
import os
from IPython.display import Image, display

def save_clean_led_circuit():
    output_path = os.path.join(os.getcwd(), 'clean_rpi_led.png')

    with schemdraw.Drawing(show=False) as d:
        d.config(unit=3)

        bb = pictorial.Breadboard().at((0, -3))
        d += bb

        # Raspberry Pi label
        d += elm.Label().label('Raspberry Pi', loc='top').at((-6, 0))

        # GPIO 18
        d += elm.Dot().at((-6, -1)).label('GPIO 18', loc='left')
        d += elm.Line().right().tox(bb.F15.x).at((-6, -1)).color('green')
        d += elm.Line().down().toy(bb.F15.y).at((bb.F15.x, -1)).color('green')
        d += elm.Dot().at(bb.F15)

        # GND shifted right
        d += elm.Dot().at((-6, -2)).label('GND', loc='left')
        gnd_x, gnd_y = bb.J17.x, bb.F6.y - 4  # موقع جديد للأرضي
        d += elm.Line().right().tox(gnd_x).at((-6, -2)).color('black')
        d += elm.Line().down().toy(gnd_y).at((gnd_x, -2)).color('black')
        #d += elm.Ground().at((gnd_x, gnd_y))

        # Resistor
        resistor = pictorial.Resistor(330).at(bb.F15).to(bb.F17)
        #d += resistor.label('R=330Ω', loc='bottom')

        # LED
        led = pictorial.LED(lead_length=0.3 * pictorial.INCH).at(bb.G17)
        #d += led.label('LED', loc='right')

        # Cathode to new ground rail
        d += elm.Line().right().at(bb.H17).tox(bb.J17).color('black')
        #d += elm.Line().down().toy(gnd_y).at((bb.J17.x, bb.J17.y)).color('black')
        d += elm.Dot().at(bb.J17)

        # Dotted ground rail connection
        #d += elm.Line().at((gnd_x, gnd_y)).to(bb.J17).color('black').linestyle('--')
      #  d += elm.Label().at((gnd_x + 1, gnd_y)).label('Ground Rail', loc='left')

        d.save(output_path, dpi=300)

    display(Image(output_path))

save_clean_led_circuit()
```
